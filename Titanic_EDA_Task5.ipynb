```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r"C:\Users\RAVI OJHA\Downloads\Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
#calculate % of null values in age column

null_percentage = df['Age'].isnull().mean() * 100
print(f"Percentage of missing values in Age column: {null_percentage:.2f}%")
```

```python
#show null values in heatmap

plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis", yticklabels=False)
plt.title("Heatmap of Missing Values")
plt.show()
```

```python
#fill null values by median
age_meadian=df["Age"].median()
df["Age"].fillna(age_meadian,inplace=True)
df["Age"]
```

```python
#change age column float to int

df["Age"]=df["Age"].round().astype(int)
```

```python
df["Age"].unique()
```

```python
df.describe()
```

```python
#check duplicate values

df.duplicated().sum()
```

```python
df.head()
```

```python
#how many passenger survives class wise

df[["Survived","Pclass"]].value_counts()
```

```python
#show class wise survived by stacked plot

crosstab=pd.crosstab(df["Pclass"],df["Survived"])
crosstab.plot(kind="bar",stacked=True,colormap="viridis")
plt.title("Class wise survived")
plt.xlabel("Passenger class")
plt.ylabel("Number of passenger")
plt.legend(title="Survived")
plt.show()
```

```python
#male vs female survived


survived = df.groupby(['Sex', 'Survived']).size().unstack()
survived
```

```python
#male vs female survived by stacked chart

plt.figure(figsize=[5,5])
survived.plot(kind="bar",stacked=True,colormap="viridis")
plt.title("Male Vs Female Survived")
plt.xlabel("Sex")
plt.ylabel("number of passenger")
for container in plt.gca().containers:
    plt.bar_label(container)
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#how many passenger travles in different class
df["Pclass"].value_counts()
```

```python
#how many passenger in every class

lo=sns.countplot(x="Pclass",data=df)
plt.title("Passenger and his class")
plt.xlabel("Passenger Class")
plt.ylabel("count no of passengers")
for i in lo.containers:
    plt.bar_label(i)
plt.show()
```

```python
df["Sex"].value_counts()
```

```python
#What is the gender distribution and survival rate by gender?


plt.figure(figsize=[5,5])
gender=sns.countplot(x="Sex",hue="Survived",data=df)
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.title("Survival by Gender")
for i in gender.containers:
    plt.bar_label(i)
plt.show()
```

```python


mo=sns.countplot(x='Embarked', data=df)
plt.title("Number of Passengers by Embarkation Port")
plt.xlabel("Embarked Port")
plt.ylabel("Passenger Count")
for i in mo.containers:
    plt.bar_label(i)
plt.show()
```

```python
#survived percentage

df["Survived"].value_counts().plot(kind="pie",autopct="%.2f")
plt.show()
```

```python
#Distribution by age

plt.figure(figsize=(8,5))
sns.histplot(df['Age'].dropna(), kde=True, bins=30, color='skyblue')
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#fare distribution

sns.histplot(df['Fare'], kde=True, bins=40, color='orange')
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()
```

```python
#How does passenger class affect survival?

jo=sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival by Passenger Class")
plt.xlabel("Passenger class")
for i in jo.containers:
    plt.bar_label(i)
plt.show()
```

```python
#Does age affect survival?

zo=sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()
```

```python
#what are the correlation between numerical variables?


sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Columns)")
plt.show()
```

```python
#show all pariwise relationships

sns.pairplot(df[['Survived', 'Pclass', 'Age', 'Fare']], hue='Survived')
```

```python
#survival by embarked port and gender

sns.catplot(x="Embarked", hue="Sex", col="Survived", data=df, kind="count")
```

```python
#Box plot for fare column


plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Fare'], color='skyblue')
plt.title("Boxplot of Fare")
plt.xlabel("Fare")
plt.show()
```

```python
#Fare vs Pclass 

plt.figure(figsize=(8,5))
sns.boxplot(x='Pclass', y='Fare', data=df, palette='pastel')
plt.title("Fare Distribution across Passenger Classes")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")
plt.show()
```

```python
#average fare per class
df.groupby('Pclass')['Fare'].mean()
```

```python
#Passenger count per class
df['Pclass'].value_counts()
```

```python
#Average Fare by Class
avg_fare = df.groupby('Pclass')['Fare'].mean().reset_index()

sns.barplot(x='Pclass', y='Fare', data=avg_fare, palette='cool')
plt.title("Average Fare by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Average Fare")
plt.show()
```
